# Ação humana em fraudes de cartão de crédito

Criar um algoritmo que faça uma predição probabilistica com os dados de fraude de cartão de crédito e retorne os dados que possuem uma chance de 50% à 70% de ser fraude.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df_credit_card = pd.read_csv('creditcard.csv')
df_credit_card.head()
```

```python
x,y = df_credit_card.drop(['Class'],axis= 'columns'),df_credit_card.iloc[: ,[30]]
```

```python

df1 = df_credit_card[['Class','V1','V2','V3','V4','V5']]
df1
```

```python
df_credit_card.info()
```

```python
df_credit_card.Class.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import  LogisticRegression
```

```python
X_treino, X_teste, Y_treino, Y_teste = train_test_split(x, y.values, test_size=0.3, random_state=12)
```

```python
lr = LogisticRegression(solver='liblinear',random_state=42)
```

```python
lr.fit(X_treino ,Y_treino)
```

```python
predicoes = lr.predict_proba(X_teste)
```

```python
predicoes[: ,1]
```

```python
valores =[]

for pre in predicoes[: ,1]:
  if pre >= 0.5 and pre <= 0.7 :
      valores.append(pre)
       


```

```python
final =pd.DataFrame(valores)
```

```python
df_credit_card['Previsoes entre 50 e 70 %'] = final
df_credit_card.head(20)
```
